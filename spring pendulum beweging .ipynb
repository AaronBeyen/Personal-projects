```python
g = 9.81
l = 1
k = 1
m = 1

# P = [x, x', theta, theta'] met ' = afgeleide en P[0] = x

def dP_dt(P, t):
    return [P[1], (l+P[0])*P[3]**2-k/m *P[0]+g*np.cos(P[2]),  P[3], -g/(l+P[0])*np.sin(P[2]) -2*P[1]/(l+P[0])*P[3]] # in de vergelijking dy/dt = f(y,t) is dit f(y,t)
```

```python
ts =  np.linspace(0, 10, 500)


x0 = 10
v0 = 0
theta_0 = np.pi/2
omega_0 = 0

P0 = [x0, v0, theta_0, omega_0]
Ps = odeint(dP_dt, P0, ts)
y0 = Ps[:,0] # x(t)
y1 = Ps[:,1] # v(t) = d x/dt
y2 = Ps[:,2] # theta(t)
y3 = Ps[:,3] # omega(t)
```

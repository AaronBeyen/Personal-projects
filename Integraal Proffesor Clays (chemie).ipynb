```python
import sympy as sp
import numpy as np
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from scipy import special

from IPython.display import display
sp.init_printing()
%matplotlib notebook
```

```python
x = sp.symbols('x')

x_vals = np.linspace(0, 5*1.6*10**(-19), 5000)

k = 1.38064852 * 10**(-23)

kleur = ['k', 'b', 'y', 'cyan', 'r', 'g', 'orange', 'grey', 'purple', 'pink']
legende = []

for j in range(15):
    fig, ax = plt.subplots(nrows=1, ncols=1, dpi=150, figsize=(8, 4))
    T = 0.1 + (100 * j)
    if T == 0:
        plt.plot(x_vals, np.linspace(0,0,5000), label = 'T = ' + str(T))
        plt.plot(x_vals, 2/np.sqrt(np.pi) *1/np.sqrt(k*T) * np.exp(-x_vals/(k*T)), label = 'oude theorie + T = ' + str(T))
    else:
        plt.plot(x_vals, 1 - special.erf(np.sqrt(x_vals/(k*T))) + 2/np.sqrt(np.pi) *np.sqrt(x_vals)/np.sqrt(k*T)* np.exp(-x_vals/(k*T)) , label = 'T = ' + str(T))
        plt.plot(x_vals, 2/np.sqrt(np.pi) *1/np.sqrt(k*T)* np.exp(-x_vals/(k*T)) , label = 'oude theorie + T = ' + str(T))
    ax.set_title('erf')
    ax.grid(True, which = 'both') #Voegt rijen en kolommen toe, zodat de datapunten beter vergeleken en afgelezen kunnen worden.
    plt.ylabel('f(x) = erf', size=10)
    plt.xlabel('$x$', size=10)
    ax.legend()
    plt.tick_params(axis='both', which='major', labelsize=10)
    ax.set_ylim(0,2)
    plt.show()
```

```python
x = sp.symbols('x')

x_vals = np.linspace(0, 0.3*1.6*10**(-19), 5000)

k = 1.38064852 * 10**(-23)

kleur = ['k', 'b', 'y', 'cyan', 'r', 'g', 'orange', 'grey', 'purple', 'pink', 'brown']
legende = []

fig, ax = plt.subplots(nrows=1, ncols=1, dpi=150, figsize=(8, 6))

for j in range(len(kleur)):
    T = 0 + (100 * j)
    if T == 0:
        plt.plot(x_vals, np.linspace(0,0,5000), label = 'T = ' + str(T))
    else:
        plt.plot(x_vals, 1 - special.erf(np.sqrt(x_vals/(k*T))) + 2/np.sqrt(np.pi) *np.sqrt(x_vals)/np.sqrt(k*T)* np.exp(-x_vals/(k*T)) , color = kleur[j], label = 'T = ' + str(T))
    ax.set_title('Fractie deeltjes voldoende energie in functie van $E_{act}$ bij verschillende temperaturen T')
    ax.grid(True, which = 'both') #Voegt rijen en kolommen toe, zodat de datapunten beter vergeleken en afgelezen kunnen worden.
    plt.ylabel('$N_{E_{act} > E_{kin}}$/$N_{tot}$', size=10)
    plt.xlabel('$E_{act}$ (J)', size=10)
    ax.legend()
    plt.tick_params(axis='both', which='major', labelsize=10)
    ax.set_ylim(0,1.1)
    ax.set_xlim(0, 0.2*1.6*10**(-19))
plt.show()
```

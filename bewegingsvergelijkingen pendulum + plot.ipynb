```python
import sympy as sp
import numpy as np
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from scipy.integrate import odeint

from IPython.display import display
sp.init_printing()
%matplotlib notebook
```

```python
g = 9.81
l = 1
b = 0


# P = [theta, theata'] met ' = afgeleide

def dP_dt(P, t):
    return [P[1], -b*P[1]-g/l*np.sin(P[0])] # in de vergelijking dy/dt = f(y,t) is dit f(y,t)
```

```python
ts =  np.linspace(0, 10, 500)

theta_0 = np.pi/3
omega_0 = 0

P0 = [theta_0, omega_0]
Ps = odeint(dP_dt, P0, ts)
y0 = Ps[:,0] # theta(t)
y1 = Ps[:,1] # omega = d theta/dt
```

```python
fig, ax = plt.subplots(1,1)
ax.set_xlim([0, 10])
ax.set_ylim([-180, 180])

# Vul hieronder de code aan.
X = ts
Y = y0
Y = Y*180/np.pi
Z = y1
K = theta_0 * np.cos(np.sqrt(g/l)*X) + omega_0 * np.sqrt(l/g) * np.sin(np.sqrt(g/l)*X) # kleine hoek benadering
K = K*180/np.pi
functie, = ax.plot([], [])
functie2, = ax.plot([], [])
dot, = ax.plot([], [], 'o', color='red')
dot2, = ax.plot([], [], 'o', color='green')

def animate(i):
    functie.set_data(X[:i], Y[:i])
    dot.set_data(X[i],Y[i])
    functie2.set_data(X[:i], K[:i])
    dot2.set_data(X[i],K[i])
plt.grid(True)

anim = animation.FuncAnimation(fig, animate, frames=len(X), interval=10)
plt.show()
```
